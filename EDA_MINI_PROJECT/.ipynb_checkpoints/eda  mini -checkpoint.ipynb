```python
#Importing all the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Reading the file.
f = pd.read_csv('fifa.csv')
```

### Converstions

```python
#Converting for CVU and Release Clause

def convert(x):
    a  = x.replace(['K','M','\u20ac'],'',regex = True).astype(float) * x.str.extract('([KM])',expand = False).replace(['K','M'],[10**3,10**6])
    return a
f['Release Clause'] = convert(f['Release Clause'])
```

```python
#explaination of the above cell.
#We need to replace k and m with million , so the way we go about with this is by first replacing k,m and the euro sign with nothing
#Note = We put regex = True when we try to replace a string with a string otherwise it would replace
#We use extract to extract k and m and the replace k and m with 10**3,10**6
```

```python
f['Release Clause']
```

```python
#Converting Wage
f['Wage'] = convert(f['Wage'])
```

```python
f['Wage']
```

```python
#Converting Value
f['Value'] = convert(f['Value'])
```

```python
f['Value']
```

```python
#Converting Weight and height
f['Height'] = f['Height'].replace("'",'.',regex=True).astype(float)
f['Height'] = f['Height'].fillna(f['Height'].mean())




```

```python
#Converting Height from feet and inches to cm
l = []# We first create an empty list
for n in f['Height']:#Use a for loop to read each element in the series
    h,i = str(n).split('.')# We convert n into a str because we converted it into float in the above cell and float doesnt have an attribute called split
    h = float(h)*30.48#We convert it back into float since we need to mulitply
    i = float(i)*2.54
    tot = h+i
    l.append(round(tot,2))#Append the total into a list and we use round function to get upto two values after the decimal point
f['HeightCm'] = pd.Series(l)#Coverting the list into series to put it into the data frame named as a certain column
f
```

```python
#Removing suffix lbs from column weight
f['Weight'] = f['Weight'].replace('lbs',"", regex = True).astype(float)
f['Weight']
```

```python
#Converting all the cells in column Contract Valid Until into years
f['Contract Valid Until'] = pd.DatetimeIndex(f['Contract Valid Until'] ).year
f['Contract Valid Until'] 
```

```python
#Dropping columns which we do not require
f.drop(columns = ['Photo','Flag','Club Logo','Loaned From','Joined','Jersey Number'],inplace = True,axis = 1)
f
```

```python
#Checking for null values and replacing them with mean values
lis = ['ID', 'Name', 'Age', 'Nationality', 'Overall', 'Potential', 'Club',
       'Value', 'Wage', 'Preferred Foot', 'International Reputation',
       'Weak Foot', 'Skill Moves', 'Work Rate', 'Body Type', 'Position',
       'Contract Valid Until', 'Height', 'Weight']#First we creat a list of columns so reading of the columns can be done in one shot
for e in lis:#We create a loop that reads the value in the list and we place the values of the list in identifier e
    g =pd.isnull(f[e]).value_counts()#Here we use the pandas isnull func to check if the there are null values 
    #and this is in the form of a series of boolean values and we use value_counts() so as to count the number of True or False that exists in the particular column.
    print(g)
```

```python
lis2 = ['Crossing', 'Finishing',
       'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling', 'Curve',
       'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes', 'Release Clause',
       'HeightCm']
for o in lis2:
    k = pd.isnull(f[o]).value_counts()
    print(k)
```

```python
#To replace all the Null values with mean()
lis3 = ['Potential',
       'Value', 'Wage','International Reputation',
       'Weak Foot', 'Skill Moves',
       'Height', 'Weight','Crossing', 'Finishing',
       'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling', 'Curve',
       'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes', 'Release Clause',
       ]
for y in lis3:
    f[y] = f[y].fillna(f[y].mean())
```

```python
for o in f.columns:
    k = pd.isnull(f[o]).value_counts()
    print(k)
```

```python
f['Contract Valid Until'] = f['Contract Valid Until'].fillna(f['Contract Valid Until'].mean())
f['Contract Valid Until'] 
```

```python
f['Contract Valid Until'] = f['Contract Valid Until'].astype(int)
```

```python
#Converting categorical values into Numerical Values
f['Body Type'] = f['Body Type'].replace(['Lean','Normal',"Stocky"],[1,2,3])#We are replacing categorical values That is Strings to numerical values
#Since we need to fill up the nan values with mean and on top o that the categorical values are repeated so it makes it easier to replace with numerical values
```

```python
f['Body Type'] = f['Body Type'].fillna(f['Body Type'].mean()).astype(int)#We fill the nan values with the mean value
```

```python
f['Body Type']
```

```python
#converting work rate into categorical values
#h/h = 1.1
#h/m = 1.2
#h/l = 1.3
#m/h = 2.1
#m/m = 2.2
#m/l = 2.3
#l/h = 3.1
#l/m = 3.2
#l/l = 3.3
f['Work Rate'] = f['Work Rate'].replace(to_replace = ['High/ High','High/ Medium','High/ Low','Medium/ High','Medium/ Medium','Medium/ Low','Low/ High','Low/ Medium','Low/ Low'],value = [1.1,1.2,1.3,2.1,2.2,2.3,3.1,3.2,3.3])
```

```python
r = round(f['Work Rate'].mean())
d = f['Work Rate'].fillna(r)
f['Work Rate']=d

f['Work Rate']
```

```python
#replaceing work rate with a consistent mean value(some manipluation was done here)
f['Work Rate'] = f['Work Rate'].replace(2,2.1)
```

```python
f['Preferred Foot'] = f['Preferred Foot'].replace(['Left','Right'],[1,2])
```

```python
q = round(f['Preferred Foot'].mean())
f['Preferred Foot'] = f['Preferred Foot'].fillna(q)
```

```python
#dropping rows based on null values present in club
f = f[pd.notnull(f['Club'])]
f
```

```python
#Dropping rows based on the null values present in Position
f = f[pd.notnull(f['Position'])]
f
```

```python
for n in f.columns:
    x = pd.isnull(f[n]).value_counts()
    print(x)
```
